```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
print(df.isnull().sum())
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.catplot(x='Pclass', hue='Survived', col='Sex',
            data=df, kind='count', height=4, aspect=0.7)
plt.subplots_adjust(top=0.8)
plt.suptitle('Survival by Class and Sex')
plt.show()
```

```python

df_numeric = df.copy()
df_numeric['Sex'] = df_numeric['Sex'].map({'male': 0, 'female': 1})
df_numeric['Embarked'] = df_numeric['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
df_numeric = df_numeric.drop(['Name', 'Ticket', 'Cabin'], axis=1)
corr = df_numeric.corr()
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```
